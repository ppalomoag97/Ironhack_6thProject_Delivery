```python

df_orders = pd.read_csv('/home/ubuntu/Ironhack_all/Ironhack_6thProject_Delivery/project_dataset/sql_tables_sample/orders.csv', delimiter=';')
df_courier = pd.read_csv('/home/ubuntu/Ironhack_all/Ironhack_6thProject_Delivery/project_dataset/sql_tables_sample/customer_courier_chat_messages.csv', delimiter = ';')
```

```python
df_courier = df_courier.rename(columns={'chat_started_by_message': 'chat_started_by'})
df_courier.insert(3,'Message',['Give me food','Coming','Fooood','Where do you live'])
```

```python
df_courier['message_sent_time'] = pd.to_datetime(df_courier['message_sent_time'], format='%d/%m/%Y %H:%M')

df_courier
```

```python
connector = sqlite3.connect(':memory:')
df_courier.to_sql('df_courier',connector,index=False, if_exists='replace')
```

```python
query2 = '''
select *
from df_courier
'''

query2_result = pd.read_sql_query(query2,connector)

query2_result
```

```python
query = '''
with elapsed_time as
(select order_id as order_number, elapsed_seconds
from
(select order_id, message_sent_time, 
lag(message_sent_time) over (partition by order_id order by message_sent_time asc) as previous_message_timestamp,
row_number () over (partition by order_id order by message_sent_time asc) as rn,
strftime('%s',message_sent_time) - strftime('%s',lag(message_sent_time) over (partition by order_id order by message_sent_time asc)) as elapsed_seconds
from df_courier)
where rn=2),

last_order as
(select order_id2, order_stage
from
(select order_id as order_id2, row_number() over (partition by order_id order by message_sent_time desc) as rn, order_stage
from df_courier)
where rn = 1)

select order_id, 
(select min(message_sent_time)
from df_courier
where order_id = a.order_id and sender_app_type like 'Courier%') as first_courier_message_timestamp,
(select min(message_sent_time)
from df_courier
where order_id = a.order_id and sender_app_type like 'Customer%') as first_customer_message_timestamp,
sum
    (case when sender_app_type like '%Courier%' then 1
    else 0
    end) 
as courier_messages,
sum
    (case when sender_app_type like '%Customer%' then 1
    else 0
    end) 
as customer_messages, 
max
    (case when sender_app_type like '%Customer%' and chat_started_by = 't' then 1
    when sender_app_type like '%Courier%' and chat_started_by = 't' then 2
    else null
    end) as first_message_sender,
min(message_sent_time) as first_message_timestamp,
elapsed_seconds as time_elapsed_until_first_response,
max(message_sent_time) as last_message_timestamp,
c.order_stage
from df_courier a 
left join elapsed_time b
on a.order_id = b.order_number
left join last_order c
on a.order_id = c.order_id2
group by order_id
'''
result = pd.read_sql_query(query,connector)
```

***4. Exploratory Data Analysis (Statistical Programming Exercise) (25 points)***

```python
df_fake_orders.head(5)
```

```python
df_fake_orders['price_difference'] = df_fake_orders['purchase_total_price'] - df_fake_orders['products_total']
df_fake_orders['under_authorized'] = df_fake_orders['price_difference'] > 0
```

```python
df_fake_orders[df_fake_orders['final_status'] != 'CanceledStatus']

round(df_fake_orders['under_authorized'].sum() / df_fake_orders['under_authorized'].count(),2)

# 1) Percentage of under_authorized orders
```

```python
df_fake_orders['new_checkout_price'] = df_fake_orders['products_total'] *1.2

round(((df_fake_orders['purchase_total_price'] - df_fake_orders['new_checkout_price']) > 0).sum() / df_fake_orders['under_authorized'].count(),2)

# 2) What percent of orders would be correctly authorized with incremental authorization (+20%) on the amount at checkout? 
```

```python
df_fake_orders['new_under_authorized'] = (df_fake_orders['purchase_total_price'] - df_fake_orders['new_checkout_price']) > 0

df_groupped = df_fake_orders.groupby('country_code').agg({'new_under_authorized':'sum', 'order_id':'count'}).reset_index()
df_groupped.rename(columns={'new_under_authorized' : 'under_authorized_orders', 'order_id': 'total_orders'}, inplace=True)

df_groupped['percentage_under_authorized'] = (df_groupped['under_authorized_orders']/df_groupped['total_orders'])

df_groupped = df_groupped.sort_values(by='percentage_under_authorized', ascending=False)

df_groupped.head()
# 3) Are there differences when split by country?
```

```python
import matplotlib.pyplot as plt
import numpy as np

plt.figure(figsize=(12, 5))
width = 0.35
index = np.arange(len(df_groupped['country_code']))
plt.bar(index - width/2, df_groupped['under_authorized_orders'],width, color='skyblue')
plt.bar(index + width/2, df_groupped['total_orders'],width, color='orange')
plt.xlabel('Countries')
plt.ylabel('Under authorized')
plt.xticks(index - width/2, df_groupped['country_code'])
plt.show()


```

```python
plt.figure(figsize=(12,6))
plt.bar(df_groupped['country_code'],round(df_groupped['percentage_under_authorized'],2),color='maroon')
plt.xlabel('Countries')
plt.ylabel('Percentage Under Authorized')

plt.show()
```

```python
#4) For the remainder of orders that would be outside of incremental authorization, what values would be necessary to capture the remaining amount?

df_fake_orders['new_price_diff'] = df_fake_orders['purchase_total_price'] - df_fake_orders['new_checkout_price']

max_value = df_fake_orders['new_price_diff'].max()

max_row = df_fake_orders[df_fake_orders['new_price_diff'] == max_value]

required_increase = (max_row['purchase_total_price'] - max_row['new_checkout_price'])/max_row['new_checkout_price']

required_increase


```

```python
# 5) Which stores are the most problematic in terms of orders and monetary value?
df_fake_orders['status_count'] = df_fake_orders['final_status'].apply(lambda x: 1 if x=='CanceledStatus' else 0)
df_fake_orders['abs_new_price_diff'] = abs(df_fake_orders['new_price_diff'])
problematic_stores = df_fake_orders.groupby('store_address').agg({'status_count': 'sum', 'abs_new_price_diff': 'max'})

problematic_stores.rename(columns= {'status_count':'Cancelled orders'}, inplace=True)

print(problematic_stores.sort_values(by='Cancelled orders', ascending=False).head(5))

print(problematic_stores.sort_values(by='abs_new_price_diff', ascending=False).head(5))
```
